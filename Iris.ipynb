```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0) #X_train=75% of dataset
X_test.shape
```

```python
%matplotlib notebook
import pandas as pd
from pandas import DataFrame
iris_dataframe = DataFrame(X_train, columns=iris_dataset.feature_names)
# создаем матрицу рассеивания из dataframe, цвет точек задан с помощью y_train
grr=pd.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60) #старая функция
```

```python
%matplotlib inline
# создаем матрицу рассеивания из dataframe, цвет точек задан с помощью y_train
grr=pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10) #knn включает алгоритм, который будет использоваться для построения модели на обучающих данных
#а также алгоритм, который сгенирирует прогнозы для новых точек
#он также будет содержать инфо, которую алгоритм извлек из обучающих данных
#для построения модели вызываем метод fit, для которого аргументы X_train, а значения - y_train
knn.fit(X_train, y_train)
```

```python
#метод fit возвращает объект knn, таким образом мы получаем строковое значение нашего классификатора
import numpy as np
X_new=np.array([[5,2.9,1,0.2]])
prediction=knn.predict(X_new)
print("Прогноз:{}".format(prediction))
print("Спрогнозированная метка:{}".format(iris_dataset.target_names[prediction]))
```

```python
#проверка модели
y_pred=knn.predict(X_test)
np.mean(y_pred==y_test)
np.mean(y_test==y_pred)
```

```python
#метод score вычисляет правильность модели для тестового набора
knn.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np
iris_dataset=load_iris()
X_train, X_test, y_train, y_test =train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)

for i in range (1,111,10):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred=knn.predict(X_test)
    score = knn.score(X_test, y_test)
    print(i, score)
```
