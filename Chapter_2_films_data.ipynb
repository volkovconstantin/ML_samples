```python
import pandas as pd

unames = ['user_id', 'gender', 'age' ,'occupation', 'zip']
users=pd.read_table(r'C:\Users\Константин\Desktop\users.dat', sep='::', header=None, names=unames, engine='python')

rnames=['user_id', 'movie_id', 'rating', 'timestamp']
ratings=pd.read_table(r'C:\Users\Константин\Desktop\ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames =['movie_id', 'title', 'genres']
movies=pd.read_table(r'C:\Users\Константин\Desktop\movies.dat', sep='::', header=None, names=mnames, engine='python')

users[:3]
```

```python
mean_rating = pd.pivot_table(data, values ='rating', index='title', columns='gender', aggfunc='mean')
mean_rating[:5]
```

```python
mean_rating = pd.pivot_table(data, values ='rating', index='title', columns='gender', aggfunc='mean')
mean_rating[:5]
```

```python
ratings_by_title = data.groupby('title').size().sort_values(ascending=False)
active_titles = ratings_by_title.index[ratings_by_title>250]
```

```python
mean_rating_2=mean_rating.loc[active_titles]
mean_rating_2.sort_values(by='F', ascending=False)[::100]
```

```python
mean_rating_2['diff'] = mean_rating_2.F-mean_rating_2.M
mean_rating_2.sort_values(by='diff', ascending=False)[::-1][::100]
```
