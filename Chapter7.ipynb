```python
import pandas as pd
from pandas import DataFrame as df
df1=df({'key':['b','b','a','c','a','a','b'], 'data1':range(7)})
df2=df({'key':['a','b','d'], 'data2':range(3)})
pd.merge(df1, df2, on='key', how='outer').sort_values(by='key')
```

```python
how_method='outer'
df1.merge(df2, how=how_method)
```

```python
df3=df({'lkey':['b','b','a','c','a','a','b'], 'data1':range(7)})
df4=df({'rkey':['a','b','d'], 'data2':range(3)})
pd.merge(df3,df4, left_on='lkey', right_on='rkey')
```

```python
df5=df({'key':['b','b','a','c','a','b'], 'data1':range(6)})
df6=df({'key':['a','b','a','b','d'], 'data2':range(5)})
pd.merge(df5, df6, left_index=False,on='key', how='inner')
```

```python
left2=df([[1,2],[3,4],[5,6]], index=['a','c','e'], columns=['Ohio', 'Nevada'])
right2=df([[7,8], [9,10],[11,12],[13,14]], index=['b','c','d','e'], columns=['Missouri', 'Alabama'])
left2.join(right2, how='outer') #join is used to merge by index
```

```python
another=df([[7,8],[9,10],[11,12],[16,17]], index=['a','c','e','f'], columns=['New York','Oregon'])
left2.join([right2, another], how='outer')
```

```python
ages=[20,22,25,27,21,23,37,31,61,45,41,32] #binning and discretization
bins=[18,25,35,60,100]
cats=pd.cut(ages, bins)
cats
```

```python
cats.codes
```

```python
pd.value_counts(cats)
```

```python
pd.cut(ages, bins, right=False) #include left border instead of right
```

```python
bin_names=['Youth','YoungAdult','MiddleAged','Senior'] #passing names of the bins
cats2=pd.cut(ages, bins, labels=bin_names)
pd.value_counts(cats2)
```

```python
data6=np.random.rand(20)
pd.cut(data6,4, precision=2)
```

```python
data7=np.random.randn(1000)
cats3=pd.qcut(data7,4) #dividing data in quantiles
pd.value_counts(cats3)
```

```python
cats4=pd.qcut(data7,[0,0.1,0.5,0.9,1]) #setting your own quantiles
pd.value_counts(cats4)
```
